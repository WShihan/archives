```python
q4 = pd.Period('2020Q4',freq="Q-DEC")  #  定义季度
q4
```

```python
(q4.asfreq('b','e') - 1).asfreq('T','s') + 16*60  # 将季度转换成日，再转换为时，最后加上具体小时
```

```python
# 创建季度序列，从2019年第三季度至2021年第四季度
rng = pd.period_range('2019Q3','2021Q4',freq="Q-DEC")
Series(np.arange(len(rng)), index=rng)
```

```python
rng.asfreq('D','s')
```

```python
# 生成100个区间序列，频率为日
rng = pd.period_range('2020-01-01',periods=100, freq='D')
s = Series(np.random.randn(len(rng)), index=rng)# 生成100随机数据
s.resample('M').count()  # 采样后聚合
```

```python
rng = pd.period_range('2021-04-30',freq="T",periods=13)
s = Series(np.random.randn(len(rng)), index=rng)
s
```

```python
s.resample("5min",closed="left", label='right').max()  # 左闭右开区间
```

```python
s.resample('5min').ohlc() 
```
