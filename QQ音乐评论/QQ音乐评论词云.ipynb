```python
import pandas as pd
import matplotlib.pyplot as plt
import jieba
import jieba.analyse as analyse
import snownlp
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
import numpy as np
from PIL import Image


```

```python
df = pd.read_csv('new轨迹.csv',encoding='utf-8',names=['昵称','评论','日期'])
df['评论']
```

```python
cmt = open('comment.txt','r',encoding='utf-8').read()
freq = analyse.extract_tags(sentence=cmt ,topK=400,withWeight=True)
freq = {i[0]:i[1] for i in freq}
```

```python
mask = np.array(Image.open("jay.png"))
color = ImageColorGenerator(mask)
wd = WordCloud(
    width=500,
    height=500,
    font_path='../黑体.ttc',
    mode="RGBA",
    max_font_size=100,
    random_state=2000,
    mask=mask,
    scale=8.0,
    color_func=color,
    background_color='white',
    relative_scaling=0.5,).generate_from_frequencies(frequencies=freq)


wd.to_array()
plt.imshow(wd,interpolation="bilinear")
plt.axis("off")
plt.show()
```
