```python
import pandas as pd
from pandas import DataFrame
from pyecharts.charts import Pie
from pyecharts.render import make_snapshot
from snapshot_selenium import snapshot
```

```python
df = pd.read_csv('new轨迹.csv', names="昵称 评论 日期".split(' ')).ffill()
df['日期'] = pd.to_datetime(df['日期'])
df = df.set_index("日期", drop=False)
df
```

```python
def func(x):
    if x >= 3 and x < 6:
        return "凌晨"
    elif x >= 6 and x < 12:
        return "上午"
    elif x >=12 and x < 14:
        return "中午"
    elif x >= 14 and x < 19:
        return "下午"
    elif x >= 19 and x<22:
        return "晚上"
    elif (x >= 22 and x <= 23) or (x >=0 and x < 3):
        return "深夜"


df['time'] = df['日期'].apply(lambda x:x.strftime('%H')).astype('int32')

df['flag'] = df['time'].apply(func)
time = df.groupby('flag').count()['time']
df[df['flag']=='深夜'][:50]
```

```python
pie = Pie()
data = time.to_list()
labels = "上午 下午 中午 凌晨 晚上 深夜".split(" ")
pie.add("时段分布",
        [(labels[i],data[i]) for i in range(6)],
       rosetype="area",)
pie.render_notebook()
#make_snapshot(snapshot, pie.render(),'时段_pie.png')
```
